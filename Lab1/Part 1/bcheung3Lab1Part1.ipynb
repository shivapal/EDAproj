```r
sales1<-c(12,14,16,29,30,45,19,20,16, 19, 34, 20)
sales2<-rpois(12,34)  # random numbers, Poisson distribution, mean at 34, 12 numbers
par(bg="cornsilk")
# plot line graph
plot(sales1, col="blue", type="o", ylim=c(0, 80), xlab="Month", ylab="Sales" )

title(main="Sales by Month")

lines(sales2, type="o", pch=22, lty=2, col="red")
grid(nx=NA, ny=NULL)
legend("topright", inset=.05, c("Sales1","Sales2"), fill=c("blue","red"), horiz=TRUE)
```

```r
library("ggmap")
library("maptools")
library(maps)
# Use google API to plot map
register_google(key = '') # Use google API key
visited <- c("SFO", "Chennai", "London", "Melbourne", "Lima", "Peru", "Johannesbury", "SA") #List of visited places
ll.visited <- geocode(visited) 
visit.x <- ll.visited$lon   #Get longitude of place visited
visit.y <- ll.visited$lat   #Get latitude of place visited
map("world", fill=TRUE, col="white", bg="lightblue", ylim=c(-60, 90), mar=c(0,0,0,0)) # Create a world map using library
points(visit.x,visit.y, col="red", pch=36)  #Plot the coordinates using longitude and latitude
```

Plot the US map

```r
library("ggmap")
library("maptools")
library(maps)
visited <- c("SFO", "New York", "Buffalo", "Dallas, TX") #List of place visited
ll.visited <- geocode(visited)
visit.x <- ll.visited$lon #get longitude of location
visit.y <- ll.visited$lat # get latitude of location
map("state", fill=TRUE, col=rainbow(50), bg="lightblue", mar=c(0,0,0,0)) # Create a state map
points(visit.x,visit.y, col="yellow", pch=36) #plot points using longitude and latitude corrdinates
```
